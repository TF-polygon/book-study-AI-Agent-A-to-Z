```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_classic.agents import Tool
```

```python
ddg_search = DuckDuckGoSearchRun()

tools = [
    Tool(
        name="DuckDuckGo Search",
        func=ddg_search.run,
        description="인터넷에서 정보를 추출하는 웹 검색 도구"
    )
]
```

```python
from langchain_classic.tools import WikipediaQueryRun
from langchain_classic.utilities import WikipediaAPIWrapper

wikipedia = WikipediaQueryRun(api_wrapper=WikipediaAPIWrapper())

tools.append(
    Tool(
        name="Wikipedia Web Search",
        func=wikipedia.run,
        description="위키피디아를 검색하는 도구"
    )
)
```

```python
from google import genai
GOOGLE_API_KEY = "MY_API"
client = genai.Client(api_key=GOOGLE_API_KEY)

try:
    for m in client.models.list():
        print(m) 
except Exception as e:
    print(f"모델 리스트 호출 실패: {e}")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="models/gemini-2.5-flash",
    google_api_key=GOOGLE_API_KEY,
    temperature=0.7,
    max_tokens=None,
    timeout=None,
    max_retries=2,
)

response = llm.invoke("안녕, Gemini! 파이썬으로 너를 사용하는 방법을 알려줘.")
print(response.content)
```

```python
from langchain_classic.agents import initialize_agent, AgentType

agent = initialize_agent(
    tools=tools, llm=llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    verbose=True
)

agent.run('인터넷에서 2026년 AI Computer Vision 분야의 최신 기술을 검색하고 위키피디아에서 가장 인기있는 AI 기술을 검색해서 정리해서 알려줘.')
```
